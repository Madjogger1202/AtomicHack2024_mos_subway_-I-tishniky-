```python
import csv
from similar_text import similar_text

def find_similar_row(csv_file_path, input_sentence, good_percent):
    with open(csv_file_path, errors='ignore') as csv_file:
        reader = csv.reader(csv_file, delimiter=",")
        similar_rows = []
        for row in reader:
            word = row[0]
            for input_word in input_sentence.split(" "):
                similarity = similar_text(input_word, word)
                similar_rows.append([int(reader.line_num), int(similarity)])
        similar_rows = sorted(similar_rows, key=lambda x: x[1], reverse=True)
        return (similar_rows)

csv_file_path = "/content/–ø–ø_—Å—Ç–∞–Ω—Ü–∏–∏.csv"
input_sentence = "–†–æ–∫–æ—Å—Å–æ–≤—Å–∫–æ–≥–æ"
good_percent = 85

similar_rows = find_similar_rows(csv_file_path, input_sentence, good_percent)
print("Similar rows found:")
print(str(similar_rows[0][0]-1))
```
